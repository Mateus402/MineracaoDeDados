# Explique como esse gráfico:

# Por que é interessante fazer um gráfico de dispersão nesse caso?
R:
A criação de um gráfico de dispersão é vantajosa devido à sua capacidade de proporcionar uma visualização ágil e simples da relação entre duas variáveis distintas. Esse tipo de gráfico auxilia na identificação de padrões e tendências nos dados, bem como na determinação da existência e natureza de uma relação entre as variáveis. Trata-se de uma ferramenta valiosa para análise exploratória de dados e apoio à tomada de decisões embasadas.

# Qual é a diferença entre o gráfico de linha na interpretação?
R:Um gráfico de dispersão mostra a relação entre duas variáveis quantitativas, usando pontos no gráfico. Ele ajuda a identificar padrões e tendências na relação entre as variáveis. Já um gráfico de linha é usado para representar a mudança de uma variável ao longo do tempo ou de outra escala contínua. Ele é útil para mostrar tendências temporais.

## Modifique a distribuição aleatória para distribuição normal - Pesquise a função numpy

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
idades = np.random.normal(0, 100, size=1000) #distribuição aleatória de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

# Interprete a distribuição distribuição e compare a diferença com a distribuição normal

R:A distribuição gerada para as idades segue uma distribuição normal, caracterizada por uma forma de sino. Isso indica que há uma concentração maior de valores em torno da média e uma diminuição gradual das frequências em direção às extremidades. Em comparação com a distribuição uniforme, a distribuição normal apresenta uma distribuição mais concentrada em torno de um valor central, em vez de ser igualmente distribuída em todas as faixas de idade.

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)

# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R:O gráfico de caixa representa a distribuição dos dados. O retângulo central indica o intervalo interquartil (IQR), que contém a maioria dos dados (50%). A linha dentro do retângulo representa a mediana. Os "bigodes" indicam a amplitude dos dados, excluindo os possíveis valores discrepantes (outliers).

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```

A diferença entre a Equipe C e a Equipe D pode ser visualizada observando a forma e a densidade dos violinos correspondentes a cada equipe, assim vendo que a equipe d com media entre 90 e 100, sendo que a pontuaçao da equipe D começa pelo 70 porem acaba entre 110, tanto a equipe C variando entre 60 ate em torno de 125 com mediana de 90.
