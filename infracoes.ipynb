```python
import matplotlib.pyplot as plt
```

```python
# Agrupar os dados por código de infração e contar a frequência de cada código
infractions_counts = df['Código da Infração'].value_counts()

# Exibir as infrações mais comuns
top_infractions = infractions_counts.head(10)

# Agrupar os dados por código de infração e contar a frequência de cada código
infractions_counts = df['Código da Infração'].value_counts()

# Selecionar as infrações mais comuns (top 10)
top_infractions = infractions_counts.head(10)

# Criar um gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
top_infractions.plot(kind='bar', ax=ax)

# Personalizar o gráfico
plt.xlabel('Código da Infração')
plt.ylabel('Quantidade de Ocorrências')
plt.title('Top 10 Infrações de Trânsito')

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
# Agrupar os dados por estado e código de infração e contar a frequência de cada combinação
infractions_by_state = df.groupby(['UF Infração', 'Descrição Abreviada Infração']).size().reset_index(name='Quantidade')

# Ordenar os dados pela quantidade de forma decrescente em cada estado
infractions_by_state = infractions_by_state.groupby('UF Infração').apply(lambda x: x.nlargest(5, 'Quantidade')).reset_index(drop=True)

# Plotar gráficos separados por UF
for state in infractions_by_state['UF Infração'].unique():
    state_data = infractions_by_state[infractions_by_state['UF Infração'] == state]
    
    # Criar um gráfico de barras para o estado atual
    fig, ax = plt.subplots(figsize=(8, 6))
    state_data.plot(kind='bar', x='Descrição Abreviada Infração', y='Quantidade', ax=ax)
    
    # Personalizar o gráfico
    plt.xlabel('Descrição Abreviada Infração')
    plt.ylabel('Quantidade de Ocorrências')
    plt.title(f'Top 5 Infrações mais Comuns - {state}')
    plt.xticks(rotation=90)
    plt.tight_layout()
    
    # Exibir o gráfico
    plt.show()
```

```python

# Agrupar os dados por estado e código de infração e contar a frequência de cada combinação
infractions_by_state = df.groupby(['UF Infração', 'Descrição Abreviada Infração']).size().reset_index(name='Quantidade')

# Obter a infração mais comum por estado
most_common_infractions = infractions_by_state.groupby('UF Infração')['Quantidade'].idxmax()
most_common_infractions = infractions_by_state.loc[most_common_infractions]

# Criar um gráfico de barras para a infração mais comum por estado
fig, ax = plt.subplots(figsize=(12, 6))
most_common_infractions.plot(kind='bar', x='Descrição Abreviada Infração', y='Quantidade', ax=ax)

# Personalizar o gráfico
plt.xlabel('Estado')
plt.ylabel('Quantidade de Ocorrências')
plt.title('Infração de Trânsito mais Comum por Estado')

# Exibir o gráfico
plt.show()
```

```python
df_acidentes = pd.read_csv('datatran2021.csv', sep=';', encoding='latin-1')
```

```python
df_acidentes
```

```python

# Agrupar os dados por estado e código de infração e contar a frequência de cada combinação
infractions_by_state = df_acidentes.groupby(['uf', 'causa_acidente']).size().reset_index(name='Quantidade')

# Obter a infração mais comum por estado
most_common_infractions = infractions_by_state.groupby('uf')['Quantidade'].idxmax()
most_common_infractions = infractions_by_state.loc[most_common_infractions]

# Criar um gráfico de barras para a infração mais comum por estado
fig, ax = plt.subplots(figsize=(12, 6))
most_common_infractions.plot(kind='bar', x='causa_acidente', y='Quantidade', ax=ax)

# Personalizar o gráfico
plt.xlabel('Estado')
plt.ylabel('Quantidade de Ocorrências')
plt.title('Acidentes de Trânsito mais Comum por Estado')

# Exibir o gráfico
plt.show()
```

```python
# Agrupar os dados por código de infração e contar a frequência de cada código
infractions_counts = df_acidentes['tipo_acidente'].value_counts()

# Exibir as infrações mais comuns
top_infractions = infractions_counts.head(10)

# Agrupar os dados por código de infração e contar a frequência de cada código
infractions_counts = df_acidentes['tipo_acidente'].value_counts()

# Selecionar as infrações mais comuns (top 10)
top_infractions = infractions_counts.head(10)

# Criar um gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
top_infractions.plot(kind='bar', ax=ax)

# Personalizar o gráfico
plt.xlabel('Código da Infração')
plt.ylabel('Quantidade de Ocorrências')
plt.title('Top 10 tipos de Acidentes de Trânsito no Brasil')

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
